3.8.3	Store the above data (the result of 3.2) in DataFrame and analyze districts that appear 3 or more than 3 years and print the corresponding state/UT, district, murders, and year in descending order.

```python
engine = create_engine("mysql+mysqlconnector://root:Lee123*9@localhost:3306/crimes")
```

```python
df = pd.read_sql("SELECT * FROM crimes_district_wise", engine)
```

```python
df
```

```python
query = """select stateorut, district, murder, years from (select stateorut, district, murder, years,  

row_number() over  

(partition by stateorut,years order by murder desc) rownum 

from crimes_district_wise where district<>'TOTAL') tab 

where tab.rownum=1;"""
```

```python
dfx = pd.read_sql(query, engine)
```

```python
dfx['years'] = dfx['years'].astype(int)
```

```python
dfx
```

```python
grps = dfx.groupby(['stateorut','years'])
```

```python
dic = dict(dfx['district'].value_counts()>=3)
filtered_dict = {k:v for (k,v) in dic.items() if v==True}
```

```python
lis = list(filtered_dict.keys())
```

```python
dfxgt3 = dfx[dfx['district'].isin(lis)]
```

```python
len(dfxgt3)
```

```python
grps = dfxgt3.groupby(['stateorut','district'])
```

```python
fig, axes = plt.subplots(54,1,figsize=(20,300))
for (ind, grp),ax in zip(grps,axes.flat):
    ax = sns.lineplot(data=grp,x='murder',y='years',ax=ax)
    ax.set_title(str(ind))
plt.tight_layout()
plt.show()
```
